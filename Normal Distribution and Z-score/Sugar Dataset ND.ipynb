```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy import stats
%matplotlib inline
```

## Assuming Normal sugar level  is around 100 - 125 mg/dL

```python
sugar_level_male = np.random.normal(110,scale = 8, size = 6000) #creating Normally distributed Random data
sugar_level_female = np.random.normal(100,scale = 8, size = 6000) #creating Normally distributed Random data
```

```python
allsugar = np.concatenate((sugar_level_male,sugar_level_female),axis=0) # Concating bothe male and female population data
df_sugar = pd.DataFrame(data = {'sugar_lvl(mg/dL)':allsugar}) #Creating a Dataframe
```

```python
df_sugar.head()
```

```python
# Plotting the above Dataframe
sns.distplot(df_sugar,color="Blue")
plt.xlabel("sugar_lvl(mg/dL)")
plt.ylabel("frequency of occurance")
plt.title("sugar level of people in mg/dL")
```

```python
Mean_sugar_level = df_sugar.mean()
Mean_sugar_level
```

```python
std_sugar_level = df_sugar.std()
std_sugar_level
```

```python
#Plotting mean of the sample
sample_means=[]
n = 50
for sample in range(0,1000):
    sample_value = np.random.choice(a = df_sugar['sugar_lvl(mg/dL)'],size = n)
    sample_mean= np.mean(sample_value)
    sample_means.append(sample_mean)
```

```python
sns.distplot(sample_means)
plt.title("Normal distribution of sample means of 50 people")
plt.xlabel("sugar level of sample means")
plt.ylabel("freq")
```

```python
np.mean(sample_means)
```

```python
np.std(sample_means)
se =std_sugar_level/np.sqrt(50)
se
```
